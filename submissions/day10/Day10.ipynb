```python
# Explain query
spark.sql("SELECT * FROM silver_events WHERE event_type='purchase'").explain(True)

```

```python

# Partitioned table
spark.sql("""
  CREATE TABLE silver_events_part
  USING DELTA
  PARTITIONED BY (event_date, event_type)
  AS SELECT * FROM silver_events
""")
```

```python

# Optimize
spark.sql("OPTIMIZE silver_events_part ZORDER BY (user_id, product_id)")
```

```python

# Benchmark
import time
start = time.time()
spark.sql("SELECT * FROM silver_events WHERE user_id=12345").count()
print(f"Time: {time.time()-start:.2f}s")
```

```python

# Benchmark
import time
start = time.time()
spark.sql("SELECT * FROM silver_events_part WHERE user_id=12345").count()
print(f"Time: {time.time()-start:.2f}s")
```
